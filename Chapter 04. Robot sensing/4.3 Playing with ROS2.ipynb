# Laser scanners in ROS2

Let's now do something a little less theoretically intensive. We are going to have a look at a laser scanner being simulated by a professional robotic simulator, **MVSim**, within the infrastructure of ROS2.

As you know, ROS2 allows for the intercommunication of different *nodes* (executables) through a system of *topics* and *messages*. When using a simulator node like **MVSim**, this node will be responsible for handling the calculations that generate the response of the sensors, and the result of this process will be published as a message on a specific topic. Then, any node which wants to use the sensor response can access it simply by subscribing to the topic and receiving the corresponding messages.

### Launching an example

If you have a working installation of ROS2 Humble on Ubuntu, you can install MVSim with

```
sudo apt install ros-humble-mvsim
```

Let's launch a simple demo that comes pre-configured with MVSim:
```
ros2 launch mvsim demo_1robot.launch.py
```

You will see two windows open: one is the GUI of the simulator itself (MVSim), and is part of the same executable. The other window, *Rviz*, is an all-purpose visualizer for ROS2. This window is **not** part of the MVSim executable, and the only way it can display the map, the robot's pose and the laser scanner readings is because it is subscribing to the relevant topics and getting the information through the ROS2 pipeline.
